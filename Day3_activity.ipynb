```python
from scipy.stats.mstats import winsorize

#Apply winsorization to the 'qft_lot' column
data1['sqft_lot']=winsorize(data1['sqft_lot'],limits=(0,0.05))

print("Outliers in 'sqft_lot' column treated using Winsorization method(5th and 95 percentiles)")
data1.head()
```

```python
data1['sqft_above']=winsorize(data1['sqft_above'],limits=(0,0.05))

print("Outliers in 'sqft_above' column treated using Winsorization method(5th and 95 percentiles)")
data1.head()
```

```python
data1['sqft_basement']=winsorize(data1['sqft_basement'],limits=(0,0.05))

print("Outliers in 'sqft_basement' column treated using Winsorization method(5th and 95 percentiles)")
data1.head()
```

```python
data1['sqft_living']=winsorize(data1['sqft_living'],limits=(0,0.05))

print("Outliers in 'sqft_living' column treated using Winsorization method(5th and 95 percentiles)")
data1.head()
```

```python
Q1_price=data1['price'].quantile(0.25)
Q3_price=data1['price'].quantile(0.75)

upper_bound_price=Q3_price+1.5*(Q3_price-Q1_price)
lower_bound_price=Q1_price-1.5*(Q3_price-Q1_price)

data1['price']=np.where(data1['price']>upper_bound_price,upper_bound_price,data1['price'])
data1['price']=np.where(data1['price']<lower_bound_price,lower_bound_price,data1['price'])

print("Outliers in 'price' column treated with IQR method" )
data1.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
data1["sqft_living"]=minmax.fit_transform(data1[["sqft_living"]])
data1.head()
```
