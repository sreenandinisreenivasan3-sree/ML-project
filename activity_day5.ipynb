```python
len_mean=df['bill_length_mm'].mean()
df['bill_length_mm']=df['bill_length_mm'].fillna(len_mean)
```

```python
dep_mean=df['bill_depth_mm'].mean()
df['bill_depth_mm']=df['bill_depth_mm'].fillna(dep_mean)
```

```python
flip_mean=df['flipper_length_mm'].mean()
df['flipper_length_mm']=df['flipper_length_mm'].fillna(flip_mean)
```

```python
mass_mean=df['body_mass_g'].mean()
df['body_mass_g']=df['body_mass_g'].fillna(mass_mean)
```

```python
df['sex'].value_counts()
```

```python
mode1=df['sex'].mode()
mode1
```

```python
df['sex'].fillna(df['sex'].mode()[0],inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['sex']=le.fit_transform(df['sex'])
```

```python
df=pd.get_dummies(df,columns=['island'],prefix=['Island'],drop_first=True)
```

```python
df.head()
```

```python
bool_col=df.select_dtypes(include='bool').columns
df[bool_col]=df[bool_col].astype(int)
```

```python
y=df.species
x=df.drop(columns=['species'])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=100)
lr.fit(x_train,y_train)
```

```python
lr.get_params()
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(random_state=42,criterion='gini',min_samples_leaf=10,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=100,min_samples_leaf=20,min_samples_split=20)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model=KNeighborsClassifier(n_neighbors=10,metric='minkowski',p=2)
knn_model.fit(x_train,y_train)
y_pred=knn_model.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```
